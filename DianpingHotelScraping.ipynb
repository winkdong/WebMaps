## Define a function to get dianping hotel main info

http://www.dianping.com/danxian/hotel + /p + 1~n

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
import pandas as pd

# http://www.dianping.com/danxian/hotel/p1
def dianping_hotel(url):
    html = urlopen(url)
    bsObj = BeautifulSoup(html,"lxml")

    n_names = bsObj.findAll("h2", {"class":"hotel-name"})
    n_links = bsObj.findAll("a", {"class":"hotel-name-link"})
    n_tags = bsObj.findAll("p", {"class":"hotel-tags"})
    n_prices = bsObj.findAll("div", {"class":"price"})
    n_ranks = bsObj.findAll("span", {"class":re.compile("^(sml-rank-stars sml-str)")})

    def liTex(node):
        li = list()
        for item in node:
            li.append(item.get_text().strip())
        return li

    hotel_name = liTex(n_names)
    hotel_tag = liTex(n_tags)
    hotel_price = liTex(n_prices)

    hotel_rank = list()
    for rank in n_ranks:
        hotel_rank.append(rank.attrs['class'][-1:])

    hotel_link = list()
    for link in n_links:
        newPage = "http://www.dianping.com{}".format(link.attrs['href'])
        hotel_link.append(newPage)
        
    danxian_list = pd.DataFrame(
    {'name':hotel_name,
    'price':hotel_price,
    'rank':hotel_rank,
     'tag':hotel_tag,
    'link':hotel_link})
    return danxian_list
```
